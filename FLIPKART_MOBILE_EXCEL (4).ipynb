```python
plt.figure(figsize=(15,5))
sns.barplot(x=redmi_smartphone['title'],y=redmi_smartphone['avg_rating'])
plt.title('Avg Rating of Redmi SmartPhones ')
plt.xlabel('Phones')
plt.ylabel('Avg Rating')
plt.ylim(ymin=3,ymax=4.7)
plt.xticks(rotation=90)
plt.show()


# insight
# The bar plot illustrates the average ratings of various Redmi smartphones, showcasing a range of customer satisfaction levels. It identifies both high and low-rated models, indicating the need for further analysis to understand the factors contributing to these variations in customer perception.
```

```python
#Rating count of Each Redmi Phon
plt.figure(figsize=(20,8))       
sns.barplot(x=redmi_smartphone['title'],y=redmi_smartphone['ratings_count'])
plt.xticks(rotation=90)
plt.title('Rating Count of Each Model')
plt.xlabel('Phones')
plt.ylabel('Rating Count')
plt.show()

# The bar plot reveals the varying levels of user engagement for different Redmi smartphone models, with some models receiving significantly more ratings than others. Models with both high average ratings and substantial rating counts are likely to be both well-received and popular among users, presenting potential market leaders. Conversely, models with low average ratings and low rating counts may require attention to improve their market performance.
```

```python
# Create a histogram of average ratings
plt.figure(figsize=(10, 6))
plt.hist(redmi_smartphone['avg_rating'], bins=15, edgecolor='k', color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Average Ratings for Redmi Smartphones')
plt.grid(True)
plt.show()

# The histogram reveals that the distribution of average ratings for Redmi smartphones is bimodal, with two distinct peaks. This suggests that Redmi offers both highly-rated and moderately-rated smartphone models, potentially catering to diverse customer preferences. Understanding this bimodal pattern can help target specific market segments effectively.
```

```python
# Create a scatter plot between 'selling_price' and 'avg_rating'
plt.figure(figsize=(10, 6))
plt.scatter(redmi_smartphone['selling_price'], redmi_smartphone['avg_rating'], alpha=0.5, color='blue')
plt.xlabel('Selling Price')
plt.ylabel('Average Rating')
plt.title('Scatter Plot: Selling Price vs. Average Rating')
plt.grid(True)
plt.show()
# The scatter plot illustrates that there is no clear linear relationship between the selling price and average rating of Redmi smartphones, indicating that customers' satisfaction is influenced by factors beyond price. It also highlights the diversity of price points at which Redmi offers smartphones and suggests potential market segments with varying preferences for price-to-rating ratios, emphasizing the importance of catering to a wide range of customer expectations.
```

```python
# Select the numerical columns for the heatmap
numerical_columns = data[['selling_price', 'avg_rating', 'ratings_count', 'reviews_count', 'one_stars_count', 'two_stars_count', 'three_stars_count', 'four_stars_count', 'five_stars_count']]

# Calculate the correlation matrix
correlation_matrix = numerical_columns.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
# The heatmap illustrates the relationships among numerical variables in the dataset:Average Rating Independence: The "avg_rating" column shows weak correlations with other numerical attributes, indicating that average ratings are largely independent of factors like price and rating counts.
# The heatmap illustrates the relationships among numerical variables in the dataset:

# Average Rating Independence: The "avg_rating" column shows weak correlations with other numerical attributes, indicating that average ratings are largely independent of factors like price and rating counts.

# Positive Rating Count Associations: Rating count columns (e.g., "ratings_count," "five_stars_count") have moderate to strong positive correlations with each other. This suggests that products with higher overall ratings tend to receive more ratings across all star categories, particularly five-star ratings.

# Price and Ratings Independence: Selling price ("selling_price") exhibits weak correlations with both rating counts and average rating, implying that the price of Redmi smartphones doesn't strongly dictate user ratings or the number of reviews.

# Complex Rating Determinants: While the heatmap provides insights into correlations, it's essential to recognize that user ratings are influenced by multifaceted factors not captured in this analysis. Other variables, such as product features, marketing, and customer experiences, likely contribute significantly to the perceived quality and popularity of Redmi smartphones.
```

```python
# Group the data by brand and count the number of products for each brand
brand_counts = data['brand'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of  Smartphones by Brand')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
# The pie chart illustrates the distribution of smartphone brands in the dataset, revealing:

Dominant Brands: A few brands hold a significant share of the smartphone market, with larger slices in the chart, suggesting their dominant presence.

Brand Diversity: Despite the dominance of some brands, the presence of multiple slices underscores the diversity of brands available to consumers, indicating a competitive and varied market.

Market Share Opportunities: Smaller slices represent opportunities for emerging or niche brands to gain market share and cater to specific consumer preferences
```
